```python
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://es.wikipedia.org/wiki/Anexo:Planetas_extrasolares'
response = get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
table = soup.find('table', class_ = "wikitable sortable")
rows = table.find_all('tr')
```

```python
head = rows[0]
columns_soup = head.find_all('th')
```

```python
columns = []
for i in range(len(columns_soup)):
    columns.append(columns_soup[i].text.strip())
```

```python
l = []
for tr in rows:
    td = tr.find_all('td')
    print(td)
    row = [tr.text for tr in td]
    l.append(row)
```

```python
df = pd.DataFrame(l, columns = columns)
df.drop
```
